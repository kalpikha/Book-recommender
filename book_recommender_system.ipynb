# Popularity based recommender System

Hightest rated books
criteria - minimum 250 ratings

-- display top 50 books with highest avg rating with minimum 250 votings

```python
#mergings dataframes books and ratings
merged_df = ratings.merge(books, on = 'ISBN')
merged_df.head(2)
```

```python
#checking the number of ratings for each book
num_ratings = merged_df.groupby('Book-Title').count()[['Book-Rating']].reset_index()
num_ratings.rename(columns = {'Book-Rating' : 'Number-of-Ratings'}, inplace = True)
num_ratings.head()
```

```python
avg_ratings = merged_df.groupby('Book-Title').mean()[['Book-Rating']].reset_index()
avg_ratings.rename(columns = {'Book-Rating' : 'Avg-Ratings'}, inplace = True)
avg_ratings.head()
```

```python
popular_df = num_ratings.merge(avg_ratings, on = 'Book-Title')
popular_df
```

```python
popular_df.describe()
```

```python
popular_df = popular_df[popular_df['Number-of-Ratings'] > 250]
top_50 = popular_df.sort_values('Avg-Ratings', ascending = False).head(50)
top_50
```

```python
top_50 = top_50.merge(books, on = 'Book-Title')
top_50.drop_duplicates('Book-Title', inplace = True)
top_50.head()
```

```python
top_50 = top_50[['Book-Title', 'Number-of-Ratings', 'Avg-Ratings', 'Book-Author','Image-URL-M']]
top_50.head()
```

# Collaborative Filtering based

```python
#consider only those users which have given a atleast 200 amount of ratings 
#only books which has minimum of 50 ratings.
#these chosen by experiment only
```

```python
x = merged_df.groupby('User-ID').count()['Book-Rating'] > 200
filtered_users = x[x].index
```

```python
filtered_ratings = merged_df[merged_df['User-ID'].isin(filtered_users)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
filtered_books = y[y].index
```

```python
final_df = filtered_ratings[filtered_ratings['Book-Title'].isin(filtered_books)]
final_df.head()
```

```python
final_df.duplicated().sum()
```

```python
final_pt = final_df.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
final_pt.fillna(0, inplace = True)
final_pt.head()
```

```python
#each book is a vector in 810 dimensional space

#find similar vectors(books) using cosine similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity as cs
```

```python
#similarity score of each book with other books
similarity_score = cs(final_pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(final_pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == final_pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('Message in a Bottle')
```

```python
import pickle

pickle.dump(final_pt,open('final_pt.pkl','wb'))
pickle.dump(books,open('books.pkl','wb'))
pickle.dump(similarity_score,open('similarity_score.pkl','wb'))
```
